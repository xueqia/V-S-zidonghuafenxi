```python
titan_=titan.drop(["name","ticket","cabin","boat","body","home.dest"],axis=1)
titan_
```

```python
x=(titan_.sex=="male")
x.astype("int")

titan_["sex"]=x.astype("int")
titan_["sex"]
titan_.head()
```

```python
titan_.info()
titan_["age"].max()
# titan_["age"].min()
```

```python
#处理缺失值

titan_["age"]=titan_["age"].fillna(titan_["age"].mean())
titan_["fare"]=titan_["fare"].fillna(titan_["fare"].mean())
titan_.info()
```

```python
labels=titan_["embarked"].unique().tolist()
labels
```

```python
titan_["embarked"]=titan_["embarked"].apply(lambda x: labels.index(x))
titan_.info()
```

```python
#划分数据集：x：survived之外的所有列 分析的特征 y:标签 survived

y=titan_.loc[:,"survived"]
y
```

```python
x=titan_.loc[:,titan_.columns!="survived"]
x
```

```python
from sklearn.model_selection import train_test_split

Xtrain,Xtest,Ytrain,Ytest = train_test_split(x,y,test_size=0.3)

Xtrain
```

```python
#sklearn分析的三部曲：
#1. 初始化决策树对象
#2. fit样本数据
#3. 求得分

from sklearn import tree

clf=tree.DecisionTreeClassifier()
clf=clf.fit(Xtrain,Ytrain)
score=clf.score(Xtest,Ytest)

score
```

```python
clf.feature_importances_  #特征的重要性
```

```python
from sklearn.model_selection import cross_val_score
import matplotlib.pyplot as plt

tr=[]
te=[]

for i in range(10):
    
    clf=tree.DecisionTreeClassifier(random_state=25
                                   ,max_depth=i+1)
    
    clf=clf.fit(Xtrain,Ytrain)
    score_tr=clf.score(Xtrain,Ytrain)
    score_te=cross_val_score(clf,x,y,cv=10).mean()
    tr.append(score_tr)
    te.append(score_te)
    
plt.plot(range(1,11),tr,color="red",label="train")
plt.plot(range(1,11),te,color="blue",label="test")

plt.show()
```
